### Take a random vector without having values &lt;1

```python
vector1 = np.random.randint([50,90])
vector1
```

### Find an orthogonal vector of that random vector

```python
orthogonal = [5] # assuming x1 to be 5

x2 = orthogonal[0] * vector1[0]
x2 = -1 * x2
x2 = x2/vector1[1]

orthogonal.append(x2)
orthogonal
```

```python
product = np.dot(vector1 , orthogonal)
product
```

### Take a projection of the random vector on orthogonal and vice versa

```python
magnitude = np.sqrt(sum(vector1**2)) 

projection_v1 = (np.dot(orthogonal, vector1)/magnitude) * (vector1/magnitude)
```

```python
projection_v1
```

```python
orthogonal = np.array(orthogonal)
magnitude = np.sqrt(sum(orthogonal**2)) 

projection_v2 = (np.dot(vector1, orthogonal)/magnitude) * (orthogonal/magnitude)
```

```python
projection_v2
```

### Take the Iris dataset and compute the eigenvalues and eigenvectors 

```python
import sklearn
from sklearn import datasets

iris_X , iris_y = datasets.load_iris(return_X_y = True , as_frame = True)
iris_X
```

```python
temp_dataframe = iris_X.to_numpy()
dataframe = np.asmatrix(temp_dataframe)

squared_dataframe = dataframe * dataframe.T
squared_dataframe
```

```python
value , vector = eig(squared_dataframe)
print(value[0])
print(vector[0])
```
